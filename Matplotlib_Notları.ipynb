```python
x = np.linspace(0, 10, 100)             # 0-10 arası 100 eşit nokta
y = np.sin(x)                           # y = sin(x)

plt.plot(x, y, color='blue', linestyle='-', marker='*', label='sin(x)')  #label plot içinde bilgi verir
# color: çizgi rengi
# linestyle: çizgi tipi ('-', '--', ':', '-.')
# marker: veri noktalarının işareti ('o', 's', '^', '*')
# label: legend için etiket

plt.title("Sine Wave")                  # Grafik başlığı
plt.xlabel("X axis")                    # X ekseni etiketi
plt.ylabel("Y axis")                    # Y ekseni etiketi
plt.grid(True)                          # Izgara çizgileri
plt.legend()                            # Etiket kutusu
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig,axs=plt.subplots(2,1)      #2 satır 1 sütun
axs[0].plot(x,y1,label="sinx")
axs[0].set_title("Sinx Grafiği")
axs[0].legend()
axs[1].plot(x,y2,label="cosx")
axs[1].set_title("Cosx Grafiği")
axs[0].legend()

plt.tight_layout()
plt.show()
```

```python
# X ve Y ızgara koordinatları
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)

# Z: Her (X, Y) noktası için değer
Z = np.sin(np.sqrt(X**2 + Y**2))

# Görselleştirme
plt.imshow(Z, extent=[-5, 5, -5, 5], origin='lower', cmap='plasma', aspect='auto')
plt.colorbar(label="Intensity")  # Renk skalası
plt.title("2D Heatmap")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
%matplotlib widget                        
                                             #3d görüntüyü inceleyebilmemizi sağlar.
fig = plt.figure()                           # Yeni bir figür (çizim alanı) oluşturur
ax = fig.add_subplot(111, projection='3d')     # 111: 1 satır, 1 sütun, 1. subplot (tek eksen) , 3D eksen ekler

x = np.linspace(-5, 5, 50)                  # -5 ile 5 arasında 50 eşit nokta
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)                    # X ve Y için 2D grid oluşturur
Z = np.sin(np.sqrt(X**2 + Y**2))            # Z değerlerini hesaplar

#ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='red')           # 3D yüzey grafiği çizer
#ax.plot3D(X, Y, Z) #: 3D çizgi grafiği
ax.scatter3D(X, Y, Z, c=Z, cmap='inferno') # 3D nokta grafiği
#ax.contour3D(X, Y, Z, levels=50)      #3D kontur grafiği  cmap='viridis': Renk haritası (colormap) olarak 'viridis' kullanılır
                                                        # edgecolor='red': Yüzeyin kenar çizgileri kırmızı olur

ax.set_title("3D Surface Plot")                                     # Grafik başlığı
plt.show()  # Grafiği gösterir
```

```python
# 1. Veri oluşturuluyor
x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)

# 2. Grafik oluşturma
fig, ax = plt.subplots()
line, = ax.plot(x, y)                        # Virgül unutulmamalı: line, = ...

                                            # 3. Güncelleme fonksiyonu: her karede değişen kısmı yamamız gerekiyor
def update(frame):
                                             # Sinüs dalgası her karede kaydırılacak
    y = np.sin(x + frame * 0.1)  
    line.set_ydata(y)
    return line,

# 4. Animasyon nesnesi oluşturuluyor
ani = animation.FuncAnimation(fig, update, frames=100, interval=50, blit=True)
plt.show()
```
