```python
x = np.linspace(0, 10, 100)             # 0-10 aras 100 eit nokta
y = np.sin(x)                           # y = sin(x)

plt.plot(x, y, color='blue', linestyle='-', marker='*', label='sin(x)')  #label plot iinde bilgi verir
# color: izgi rengi
# linestyle: izgi tipi ('-', '--', ':', '-.')
# marker: veri noktalarnn iareti ('o', 's', '^', '*')
# label: legend iin etiket

plt.title("Sine Wave")                  # Grafik bal
plt.xlabel("X axis")                    # X ekseni etiketi
plt.ylabel("Y axis")                    # Y ekseni etiketi
plt.grid(True)                          # Izgara izgileri
plt.legend()                            # Etiket kutusu
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig,axs=plt.subplots(2,1)      #2 satr 1 stun
axs[0].plot(x,y1,label="sinx")
axs[0].set_title("Sinx Grafii")
axs[0].legend()
axs[1].plot(x,y2,label="cosx")
axs[1].set_title("Cosx Grafii")
axs[0].legend()

plt.tight_layout()
plt.show()
```

```python
# X ve Y zgara koordinatlar
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)

# Z: Her (X, Y) noktas iin deer
Z = np.sin(np.sqrt(X**2 + Y**2))

# Grselletirme
plt.imshow(Z, extent=[-5, 5, -5, 5], origin='lower', cmap='plasma', aspect='auto')
plt.colorbar(label="Intensity")  # Renk skalas
plt.title("2D Heatmap")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
# 1. Veri oluturuluyor
x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)

# 2. Grafik oluturma
fig, ax = plt.subplots()
line, = ax.plot(x, y)                        # Virgl unutulmamal: line, = ...

                                            # 3. Gncelleme fonksiyonu: her karede deien ksm yamamz gerekiyor
def update(frame):
                                             # Sins dalgas her karede kaydrlacak
    y = np.sin(x + frame * 0.1)  
    line.set_ydata(y)
    return line,

# 4. Animasyon nesnesi oluturuluyor
ani = animation.FuncAnimation(fig, update, frames=100, interval=50, blit=True)
plt.show()
```
